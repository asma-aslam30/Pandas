# Merge

```python
import pandas as pd
var1=pd.DataFrame({"a":[1,2,3,4],"b":[5,6,7,8]})
var2=pd.DataFrame({"a":[1,2,3,4],"b":[54,64,74,84]})
pd.merge(var1,var2,on='a')
```

```python
import pandas as pd
var1=pd.DataFrame({"a":[1,2,3,4],"b":[5,6,7,8]})
var2=pd.DataFrame({"a":[1,2,3,4],"b":[54,64,74,84]})
pd.merge(var2,var1,on='a')
```

```python
import pandas as pd
var1=pd.DataFrame({"a":[1,2,3,4],"b":[5,6,7,8]})
var2=pd.DataFrame({"a":[1,2,3,5],"b":[54,64,74,84]})
pd.merge(var1,var2,how="outer",indicator=True)
```

```python
import pandas as pd
var1=pd.DataFrame({"a":[1,2,3,4],"b":[5,6,7,8]})
var2=pd.DataFrame({"a":[1,2,3,5],"b":[54,64,74,84]})
pd.merge(var1,var2,how="left",indicator=True)
```

```python
import pandas as pd
var1=pd.DataFrame({"a":[1,2,3,4],"b":[5,6,7,8]})
var2=pd.DataFrame({"a":[1,2,3,5],"b":[54,64,74,84]})
pd.merge(var1,var2,left_index=True,right_index=True,suffixes=("name","id"))
```

```python
import pandas as pd
va1=pd.DataFrame({"a":[1,2,3,4],"b":[5,6,7,8]})
va2=pd.DataFrame({"a":[1,2,3,5],"b":[54,64,74,84]})
pd.concat([va1,va2],axis=1)
```

```python
va1=pd.DataFrame({"a":[1,2,3,4],"b":[5,6,7,8]})
va2=pd.DataFrame({"a":[1,2,3,5],"b":[54,64,74,84]})
pd.concat([va1,va2],ignore_index=True)
```

```python
va1=pd.DataFrame({"a":[1,2,3,4],"b":[5,6,7,8]})
va2=pd.DataFrame({"a":[1,2,3,5],"b":[54,64,74,84]})
pd.concat([va1,va2],ignore_index=False)
```
